```python
import requests 
import pandas as pd 
import sqlite3

subway_od_pair_endpoint = 'https://data.ny.gov/resource/jsu2-fbtj.json'
r=requests.get(
        url = subway_od_pair_endpoint, 
        params={
            '$select':'origin_station_complex_id,destination_station_complex_id,SUM(estimated_average_ridership)',
            '$where':"destination_station_complex_name='74-Broadway (7)/Jackson Hts-Roosevelt Av (E,F,M,R)' OR destination_station_complex_name='61 St-Woodside (7)'",
            '$group':'origin_station_complex_id,destination_station_complex_id',
            '$limit':50000
            }
    )

subway_od_pair = pd.DataFrame(r.json())
subway_od_pair

conn = sqlite3.connect('bus_locations_benchmark2.db',timeout=30)
subway_od_pair.to_sql(
    'subway_od_pair_demand', conn, if_exists='append', index=False
)

```

```python
subway_stop_endpoint = 'https://data.ny.gov/resource/39hk-dx4f.json'
subway_stops = pd.DataFrame(requests.get(subway_stop_endpoint).json())
subway_stops_manh = subway_stops.loc[(subway_stops['borough']=='M')&(subway_stops['cbd']=='TRUE')][['complex_id','stop_name','gtfs_latitude','gtfs_longitude']]

#subway_stops_manh
subway_stops_manh.to_sql(
    'manhattan_subway_metadata', conn, if_exists='append', index=False
)

conn.close()
```

```python

#subway_od_pair['transfer'] = subway_od_pair['destination_station_complex_name'].apply(find_transfers)
subway_od_pair['ridership'] = subway_od_pair['SUM_estimated_average_ridership'].astype(np.float64) / 212
dm=subway_od_pair.groupby(['origin_station_complex_id',])[['ridership']].sum().sort_values(by=['ridership'],ascending=False)#.head(50)
dm.head(50)

q70_demand_ttl = 3336
dm['ridership_p'] = dm['ridership']/dm['ridership'].sum()
dm['demand'] = dm['ridership_p']*q70_demand_ttl
dm=dm.reset_index()
dm_manh_only = subway_stops_manh[['complex_id','stop_name','gtfs_latitude','gtfs_longitude']].merge(dm, left_on='complex_id', right_on='origin_station_complex_id',how='inner') 
dm_manh_only['demand_weight'] = dm_manh_only['demand']/dm_manh_only['demand'].sum()

dm_manh_only
```

```python
#!pip install osmnx
import osmnx as ox
G = ox.graph_from_place('New York, New York', network_type='drive')

G = ox.add_edge_speeds(G)
G = ox.add_edge_travel_times(G)
ox.basic_stats(G)
```

```python
import matplotlib.pyplot as plt 
lats = dm_manh_only['gtfs_latitude'].astype(float)
longs = dm_manh_only['gtfs_longitude'].astype(float)

stops = ox.nearest_nodes(G, longs, lats)
dm_manh_only['osmnx_node_id'] = stops 

nodes, edges = ox.graph_to_gdfs(G, nodes=True)
nodes = nodes.loc[stops]
```

```python
c=[]
for n1 in stops:
    print(f'**** Calculating Route Distances: {n1} ****')
    temp = []
    for n2 in stops:
        r = [int(i) for i in ox.shortest_path(G, n1, n2, weight='length')]
        if len(r) == 1:
            l = 0
        else:
            d=ox.routing.route_to_gdf(G, r)
            l = d['length'].sum()
        temp.append(l)
    c.append(temp)

distances = pd.DataFrame(c,columns=['i','j','l'])
#distances=distances.replace(dict(zip(dm_manh_only['osmnx_node_id'],dm_manh_only['stop_name'])))
#distances_matrix = distances.pivot(columns='i',index='j',values='l')
distances
```

```python
import requests
import pandas as pd 
bus_ridership_endpoint = 'https://data.ny.gov/resource/kv7t-n8in.json'
r=requests.get(
        url = bus_ridership_endpoint, 
        params={
            '$select':'transit_timestamp,SUM(ridership),SUM(transfers)',
            '$group':'transit_timestamp',
            '$where':"bus_route='Q70+'",
            '$limit':50000
        }
    )

subway_od_pair = pd.DataFrame(r.json())
subway_od_pair
```

```python
import numpy as np 

bus_data = subway_od_pair
bus_data['hour'] = pd.to_datetime(bus_data['transit_timestamp']).dt.hour
bus_data['day'] = pd.to_datetime(bus_data['transit_timestamp']).dt.day
bus_data['dow'] = pd.to_datetime(bus_data['transit_timestamp']).dt.day_of_week
bus_data['month'] = pd.to_datetime(bus_data['transit_timestamp']).dt.month 
bus_data['year'] = pd.to_datetime(bus_data['transit_timestamp']).dt.year 

bus_data = bus_data.drop(
    columns=['transit_timestamp']
)\
.rename(
    columns = {
        'SUM_ridership':'ridership',
        'SUM_transfers':'transfers'
    }
)
bus_data['ridership'] = bus_data['ridership'].astype(np.float64)
bus_data['transfers'] = bus_data['transfers'].astype(np.float64)
grp = bus_data.groupby(['day','month','year','dow']).sum().reset_index()
only2022 = grp.loc[(grp['year']==2022)&(grp['month'] < 4)&(grp['dow'] < 5)] #get most recent ridership stats

only2022
```
